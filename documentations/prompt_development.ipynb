# Process of prompt development

## Starting with problems
Since there are thousands of ways to talk to an LLM, I decided to modify my prompting strategy based on the problems. What makes high‑quality human‑generated ideas stand out is our ability to understand psychological motivations that drive product adoption. While LLMs can generate creative concepts, they often fail to resonate emotionally in the way that strong human‑generated ideas do. To address this, I’m planning to incorporate a dedicated branch in the ideation process that prompts AI to explore emotional factors, ensuring that generated ideas are not only innovative but also deeply connected to human needs and desires.

## Mind-map foundations
Mind‑mapping is by far the most prevalent and effective way for ideation. It has been adopted by renowned designers such as David M. Kelley and Naoto Fukasawa. The core of creating a mind map lies in building connections—linking ideas, concepts, and themes in a way that fosters structured yet flexible thinking. In a mind map, ideas emerge from interconnected concepts, forming a web of relationships that guide creative exploration. The process begins with a central theme or problem statement, from which branches extend outward, each representing a related subtopic, attribute, or question. As new connections are made, the combination of different elements leads to unexpected insights—merging disciplines, reframing assumptions, or introducing unconventional perspectives. As the central theme becomes more divergent, creative ideas are formed from these connections.

### Prompting experiments for "travel"
Following this track, I decided to change my prompting strategies. After exploration, I figured out several solutions, using travel as the central theme:

**Chain‑of‑Thought approach**

We are brainstorming a mind map around the word “travel.” Follow these steps: 1.	Immediate associations (the common ones). 2.	Synonyms and related concepts from various domains (literature, pop culture, psychology, technology, etc.). 3.	Metaphorical or figurative meanings of “travel” (mental journeys, historical journeys, spiritual journeys). 4.	Less obvious connections that might feel unconventional or abstract.

After listing these, combine them into a structured mind map that shows the relationships among all these levels.

**Directed role-play or perspectives**

Imagine you are:

* A poet writing about the idea of travel
* A child experiencing travel for the first time
* An environmentalist worried about the impact of travel
* A historian reflecting on ancient travel routes
* A science‑fiction writer imagining interstellar travel
  
Provide the key concepts or phrases each persona might bring up, organized in mind‑map form.

**Metaphorical and conceptual analogies**

Consider travel as a metaphor in art, music, personal growth, technology, and spirituality. For each domain, list nuanced ways travel might be understood or expressed, organized into a mind‑map.

## Prompting for emotional depth
Having these insights in mind, I started working on prompt engineering. I first came up with prompting approaches one by one and tried to combine them to form a comprehensive prompt that guides the LLM through various phases, including generating pain points, refining them emotionally, and ensuring they address specific emotional needs or user pains.

### Prompting the LLM to identify core emotions behind the problem
Identify the core emotion(s) behind the problem—what feelings (e.g., anxiety, longing, excitement) do people experience, and why do these emotions arise?

**Using the “Five Whys”**

State the problem, then repeatedly ask “Why?” until uncovering a deep emotional insight or universal human desire. Summarize the insight.

**Focusing on the holistic user experience**

Define the user persona and fill out an empathy map: Think & Feel, See, Hear, Say & Do, Pains, Gains.

**Incorporating little twists of surprise or delight**

Identify the standard design approach or norm, pinpoint a subtle playful twist, and evaluate how that moment of delight aligns with the emotional core.

## Reminders to guide feasibility and variety
Another challenge with AI‑generated ideas is their feasibility. Novelty is valuable, yet feasibility is intertwined with idea quality—an idea too ambitious to implement lacks real‑world impact. With these insights, I embedded precautions into my prompts. Inspired by Dr. Ethan Mollick’s prompt library, I added “Reminders” designed to nudge the LLM toward generating product ideas that align with predetermined directions:

* Keep the medium intimately tied to the core concept.
* Empower the user, not complicate the experience.
* Ensure the medium is realistically achievable.
* Strive for variety—avoid repetitive defaults; AI, AR/VR, or NFC solutions are last resorts.

## Pulling the pieces together: persona, goal, workflow, reminder
Using Mollick’s library as a primary reference, I broke the prompt into four sections.

**Persona**

You are an extremely creative entrepreneur with a proven track record of developing innovative, profitable products. As my co‑founder and ideation partner, your primary mission is to empower me to generate my own high‑quality ideas. Rather than listing solutions, focus on supporting my brainstorming process by offering strategic questions, frameworks, and prompts that spark unconventional thinking. Challenge my assumptions, introduce fresh perspectives, and guide me to explore new angles—while letting me take the lead.

**Goal**

In this session we will co‑create a mind‑map focused on a single problem statement at the center. We’ll explore interconnected concepts, directions, and ideas branching out from that theme and ultimately arrive at a set of well‑defined product concepts.

**Workflow**

1.	User input → LLM refines into a standardized problem statement.
2.	Refine the scope → Ask “Why?” to step up the ladder and ensure the problem is worth solving.
3.	Create the mind‑map → Present multiple branching techniques as toggles.
4.	Iterate and challenge → AI and human alternate expansions, with the LLM posing questions or unconventional perspectives.
5.	Finalize product ideas → Concepts coalesce through intuitive interactions such as click, drag, and drop.

## Refining the prompting templates

With a functioning “UI” in my terminal, I ran several tests using the ideation tool. I wasn’t satisfied with the output. Generating a standardized template for a creative workflow is inherently challenging and required significant refinement. Below are key insights from the second round of prompt engineering.

1. Be specific in requirements

LLMs excel with specific, well‑defined language. Prompts should function like precise design specifications. Adding examples, defining edge cases, and specifying response formats improve consistency. For instance, asking for “five differentiated user profiles” proved vague. Revising to “differ significantly in background, industry, occupation, age, or daily habits” gave the model clear dimensions to vary, resulting in more diverse personas.

2. Be specific in methodology

Guiding the LLM through a step‑by‑step process yields more consistent results than loosely stating the desired outcome. When combining distant concepts, I replaced broad requests with method‑driven options such as Analogy Swapping, Inversion/Role Reversal, and Contextual Transplant. By focusing on a framework, the model produced cohesive ideas and the process became predictable.

3. Pay extreme attention to output formatting

Inconsistent JSON outputs caused frontend rendering issues. I solved this by specifying exact keys and including an example:

{
  "heading": "Concise user input (5‑7 words)",
  "explanation": "Brief concept explanation (1‑2 sentences)",
  "productDirection": "Practical next step (1‑2 sentences)"
}

This explicit structure drastically improved reliability.

4. Simplify the prompt to reduce tokens (cost)

Prompt length impacts both API cost and processing time. I trimmed prompts to essential information and tested to ensure quality remained high. Cost‑efficient prompting is a strategic priority if the tool scales.

## Conclusion

By iteratively refining my prompting strategy—starting from problem‑based prompts, layering emotional depth, embedding feasibility reminders, and meticulously structuring persona, goal, workflow, and output—I shaped a comprehensive template for AIdeator. Specificity in requirements, methodology, formatting, and token economy proved crucial. Mind‑map‑centric brainstorming guided by clear emotional drivers and grounded by practical constraints enables a human‑AI partnership that balances novelty with real‑world viability, laying a strong foundation for consistent, high‑quality idea generation in future iterations.
