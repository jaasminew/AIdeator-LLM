# AIdeator Workflow: From Problem to Product

## Phase 1: Problem Definition

1. Initial Input Collection

* User provides two inputs: target audience and problem
* System processes these inputs to generate problem statements


2. Problem Statement Generation

* System generates an initial "How might we" problem statement (Problem Statement 1)
* System creates an alternative problem statement by flipping core assumptions (Problem Statement 2)
* User selects one of the two statements or requests regeneration of either

## Phase 2: Exploration Methodologies

3. Methodology Selection (by agentic AI)

* User chooses one of three exploration methodologies:

* 1. Emotional Root Causes: Examines underlying emotional needs and motivations
* 2. Unconventional Associations: Connects the problem to unexpected domains or analogies
* 3. Imaginary Customers' Feedback: Explores diverse user perspectives and pain points

4. Concept Generation

* System generates 5-7 distinct concepts based on the selected methodology
* Each concept includes a heading, explanation, and potential product direction
* Concepts are organized into branches in a mindmap structure

## Phase 3: Branch Development

5. Concept Expansion

* User selects branches to explore further
* System generates sub-branches with more specific concept variations
* User can add their own ideas to any branch


6. Branch Management

* User can edit, delete, or modify branches
* System maintains hierarchical relationships between concepts

## Phase 4: Product Development

7. Concept Combination

* User selects multiple branches to combine into more complex product ideas
* System integrates selected concepts using approaches like: Analogy Swapping; Role Reversal; Constraint Merging; Hybridization; Layering Concepts

8. Product Refinement

* Combined concepts are transformed into detailed product descriptions
* Products include features lists and clear explanations
* User can continue to edit and refine product concepts

## Phase 5: Market Research

9. Similar Product Search

* System can search for existing products similar to generated ideas
* Searches YCombinator and ProductHunt databases (2020-2025)
* Uses NLP relevancy matching to find the top 5 most similar companies
* Results include company name, description, and relevance score

10. Final Output

* User can export final product ideas with complete feature sets
* System can generate product roadmaps and budget analysis
